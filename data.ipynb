```python
import pandas as pd 

df = pd.read_csv('better.csv')
```

```python
df.sample(10)
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_df = df.copy()

encoders = {}
for i in encoded_df.select_dtypes(include='object'):
    encoder = LabelEncoder()
    encoded_df[i] = encoder.fit_transform(encoded_df[i])
    encoders[i] = encoder

encoders
```

```python
df.sample(10)
```

```python
X = encoded_df.drop(columns='racism_exposure_score')
y = encoded_df['racism_exposure_score']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=30,test_size=0.20)
```

```python
from lazypredict.Supervised import LazyRegressor
rgr = LazyRegressor(ignore_warnings=True)
models, predictions = rgr.fit(X_train,X_test,y_train,y_test)
models
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import root_mean_squared_error, mean_squared_error, r2_score

mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2s = r2_score(y_test,y_pred)

print(f'RMSE = {rmse}')
print(f'MSE = {mse}')
print(f'R2 Score = {r2s}')
```

```python
import joblib as jb

jb.dump(encoders,'myencoders.pkl')
jb.dump(model,'mymodel.pkl')
```
